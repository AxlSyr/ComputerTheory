Show that a language is decidable iff some enumerator enumerates the language in the standard string order.

Podemos usar $M$ para construir un enumerador $N$:

$N$ = Realiza lo siguiente

1. Para $i = 1, 2, 3, ...$
2. Sea $r_i$ la cadena $i$ en orden estandar de las cadenas en $\Sigma^*$, corremos $M(r_i)$.
3. Si $M$ acepta a $r_i$, imprimimos la cadena.

$\therefore$ A language is decidable $\to$ Some enumerator enumerates the language in the standard string order.

$\therefore$ $C$ is Turing-recognizable $\to$ A decidable language $D$ exists such that $C = \{x | \exists y (\langle x, y\rangle \in D)\}$

Supongamos que $C$ es reconocible por una máquina de Turing $M$.

Construimos un decididor que reciba $\langle x, y \rangle$ para $D$ que usará $y$ para detener $M$, y usará a $M$ para saber si $x \in C$ por a lo mucho $|y|$ pasos. Si $M$ acepta en esa ocasión, el decididor también aceptará, y si $M$ rechaza o si no pasa nada en los primeros $|y|$ pasos, el decididor rechazará.

Así, para alguna $x \in  C$, sabemos que $M$ aceptará la cadena después de $n$ pasosm por lo que existe una cadena $y$, tal que $|y| \ge n$, $\langle x, y \rangle \in D$.
