**$D=\{$w | w contains an equal number of occurrences of the substrings 01 and 10$\}$.**

**Thus $101 \in D$ because 101 contains a single 01 and a single 10, but $1010 \notin D$ because 1010 contains two 10s and one 01. Show that D is a regular language.**

Notamos que las cadenas que tienen la misma cantidad de ocurrencias de $01$ que de $10$ son las cuales terminan con el mismo símbolo que con el que empiezan.

P.D. Las cadenas de $D$ terminan con el mismo símbolo que con el que empiezan.

Casos base: $x$ con longitud 1 y 2

$x$ con longitud 1 se escribe como $x = 0|1$. Notamos que hay 0 ocurrencias de la subcadena $01$ y 0 ocurrencias de la subcadena $10$.

$x$ con longitud 2 se escribe como $x = 00|01|10|11$. Notamos que las opciones donde se cumple la condición de las cadenas de $D$ empiezan y terminan con el mismo símbolo, y hay 0 ocurrencias de la subcadena $01$ y 0 ocurrencias de la subcadena $10$.

Hip. Las cadenas con longitud $n$ en $D$ terminan con el mismo símbolo que con el que empiezan.

P. D. Las cadenas con longitud $n+1$ en $D$ terminan con el mismo símbolo que con el que empiezan.

Tomamos cualquier cadena $x \in D$ que tenga una longitud de $n+1$.

Ahora definamos una subcadena $x'$ que es igual a $x$ sin su último símbolo.

Entonces $x' \in D$ o $x' \notin D$.

Si $x' \in D$, por hipótesis de inducción termina con el mísmo símbolo que el símbolo con el que empieza, ya que de otra manera no tendríamos la misma cantidad de occurrencias de las dos subcadenas. Como $x' \in D$ y $x \in D$, el único simbolo que pudimos quitarle a $x$ es el mísmo que el último símbolo de $x$.

Si $x' \notin D$, entonces $x$ no puede terminar con el mismo símbolo, ya que de hacerlo sí estaría en $D$. Y con el símbolo que se le quitó a $x$, $x$ sí debe iniciar con el mismo símbolo que con el que termina, ya que de otro modo significaría que $x'\in D$.

Entonces demostramos que las cadenas de $D$ terminan con el mismo símbolo que con el que empiezan.

Entonces podemos escribir facilmente una expresión regular que describa a $D$:

$0(0|1)^*0|1(0|1)^*1$

Como pudimos encontrar una expresión regular que describe a $D$, por la equivalencia entre expresiones regulares y $FA's$, sabemos que existe un autómata finito que lo reconoce, entonces sabemos que $D$ es un lenguaje regular.

```
def get_estados(programa, inicial, aceptacion):
    '''
    Obtiene los estados de un automata

    programa: colección de transiciones del automata
    inicial: estado inicial del autómata
    aceptacion: colección de estados de aceptación
    '''
    resultado = {inicial}
    resultado.update(aceptacion)
    resultado.update(estado for estado, simbolo, siguiente in programa)
    resultado.update(siguiente for estado, simbolo, siguiente in programa)
    return frozenset(resultado)
```

```
def get_simbol(now_state, next_state, programa):
    '''
    Obtiene el símbolo de una transición

    now_state: estado origen
    next_state: estado destino
    programa: colección de transiciones del automata
    '''
    res = []
    for state, simbol, next in programa:
        if(state == now_state and next == next_state):
            res.append(simbol)
    return '|'.join(res)
```

```
def transition_exists(now_state, next_state, programa):
    '''
    Devuelve una bandera según existe una transición en un autómata

    now_state: estado origen
    next_state: estado destino
    programa: colección de transiciones del automata
    '''
    for state, simbol, next in programa:
        if(state == now_state and next == next_state):
            return True
    return False
```

```
def fa_to_regex(automata: af.AFN):
    '''
    Algorítmo para convertir un autómata finito a una expresión regular

    automata: autómata no determinista que se busca convertir a regex

    Se implementa el método de eliminación de estados

    Paso 1:
        Si el estado inicial es un estado de aceptación o si tiene transiciones
        de entrada, añadimos un nuevo estado inicial que no sea de aceptación y
        que no tenga transiciones de entrada, y creamos una transición epsilon
        entre el nuevo estado inicial y el estado inicial original.
        
    Paso 2:
        Si hay más de un estado de aceptación o si el único estado de aceptación
        tiene transiciones de salida, agregamos un nuevo estado de aceptación,
        quitamos al resto de los estados de aceptación y agregamos una
        transición epsilon desde cada estado de aceptación original al nuevo
        estado de aceptación.
    
    Paso 3:
        Eliminamos a todos los estados que no son iniciales ni de aceptación,
        y escribimos el nuevo símbolo de la transición de acuerdo con el
        procedimiento siguiente:
        Escogemos una q en Q\{inicial, aceptacion}.
        Para cada q_1, q_2 en Q\{q} creamos una nueva transición donde
        q_1 es el estado origen, el nuevo símbolo es
        simbolo_{q_1q_2}+simbolo_{q_1q}simbolo_{qq}*simbolo_{qq_2}, y el estado
        destino es q_2.
        Eliminamos q y todas sus transiciones
    '''
    # Paso 1
    transitions = list(automata.programa())
    initial = automata.estado_inicial
    acceptance = list(automata.estados_de_aceptacion)
    states = get_estados(transitions, initial, acceptance)

  
    if any(initial == x.siguiente for x in transitions) or \
        (initial in acceptance):
        transitions.append(af.Transicion('new', '', initial))
        initial = 'new'

    states = get_estados(transitions, initial, acceptance)
    
    # Paso 2
    if len(acceptance) > 1:
        for state in acceptance:
            transitions.append(af.Transicion(state, '', 'accept'))
        acceptance = ['accept']
    elif any(acceptance[0] == x.estado for x in transitions):
        transitions.append(af.Transicion(acceptance[0], '', 'accept'))
        acceptance = ['accept']

    states = get_estados(transitions, initial, acceptance)
    
    # Paso 3
    to_remove = list(get_estados(transitions, initial, acceptance))
    to_remove.remove(initial)
    to_remove.remove(acceptance[0])
    print(to_remove)
    print()


    for t in transitions:
        print(t)
    print()

    while len(to_remove) > 0:
        state = to_remove.pop(0)

        states1 = list(get_estados(transitions, initial, acceptance))
        states1.remove(state)
        remove_transitions = list(itertools.product(states1, states1))
        print(states1)
        print(state)

        for now, then in remove_transitions:
            if transition_exists(now, state, transitions) and \
                transition_exists(state, then, transitions):
                print(now + ' -> ' + then)
                part1 = get_simbol(now, then, transitions)
                part2 = ''
                if len(get_simbol(state, state, transitions)) > 0:
                    part2 += get_simbol(now, state, transitions) + '('
                    part2 += get_simbol(state, state, transitions)+ ')*'
                    part2 += get_simbol(state, then, transitions)
                else:
                    part2 += get_simbol(now, state, transitions)
                    part2 += get_simbol(state, then, transitions)

                if len(part1) > 0 and len(part2) > 0:
                    regex = part1 + '|' + part2
                else:
                    regex = part1 + part2

                if transition_exists(now, then, transitions):
                    transitions.remove(
                        af.Transicion(now, get_simbol(now, then, transitions), \
                            then))
                print(regex)
                transitions.append(af.Transicion(now, regex, then))
            else:
                continue

    
        for t in transitions[:]:
            if (t.estado == state) or (t.siguiente == state):
                transitions.remove(t)

        transitions = set(transitions)
        transitions = list(transitions)

        to_remove = list(get_estados(transitions, initial, acceptance))
        to_remove.remove(initial)
        to_remove.remove(acceptance[0])


        for t in transitions:
            print(t)
        print()

    return get_simbol(initial, acceptance[0], transitions)
```

# Un programa que convierte cualquier gramática libre de contexto en un autómata de pila

```
def cfg_to_pda(program, terminals, non_terminals, initial):
    '''
    Algorítmo para convertir una gramática libre de contexto a un autómata de
    pila

    program: producciones de la gramática
    terminals: conjunto de símbolos terminales
    non_terminals: conjunto de símbolos no terminales
    initial: símbolo inicial de la gramática

    La idea es que el procesamiento de una cadena imite la generación de una
    derivación. Como parte de una derivación, normalmente llevamos a cabo una
    sustitución de un no terminal por una cadena de símbolos de acuerdo con
    alguna regla de producción. La pila nos permite recordar qué símbolos de
    terminal tenemos que leer más abajo en la cadena para que se haya aplicado
    alguna regla de producción específica para generar la cadena dada.
    '''
    p = [
        (0, '', '', 1, '$'),
        (1, '', '', 2, initial)
    ]
    states = 2

    for simbol in program:
        prods = program[simbol]
        for string in prods:
            for char in reversed(string):
                if char == string[-1]:
                    new = (2, '', simbol, states + 1, char)
                    states += 1
                elif char == string[0]:
                    new = (states, '', '', 2, char)
                else:
                    new = (states, '', '', states + 1, char)
                    states += 1
                if len(string) == 1:
                    new = (2, new[1], new[2], 2, new[4])
                    states -= 1
                p.append(new)
    for t in terminals:
        p.append((2, t, t, 2, ''))
    p.append((2, '','$',states+1,''))
    return (p, 0, [states + 1])
```
